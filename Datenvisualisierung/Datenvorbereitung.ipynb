```python
import matplotlib.pyplot as pyplot

import pandas as pandas

df1 = pandas.read_csv("./stack-overflow-developer-survey-2021/survey_results_public.csv",index_col="ResponseId")
df2 = pandas.read_csv("./stack-overflow-developer-survey-2020/survey_results_public.csv",index_col="Respondent")
df3 = pandas.read_csv("./stack-overflow-developer-survey-2019/survey_results_public.csv",index_col="Respondent")
df4 = pandas.read_csv("./stack-overflow-developer-survey-2018/survey_results_public.csv",index_col="Respondent")
df5 = pandas.read_csv("./stack-overflow-developer-survey-2017/survey_results_public.csv",index_col="Respondent")
```

```python
list_of_dfs = list()
list_of_dfs.append(df1)
list_of_dfs.append(df2)
list_of_dfs.append(df3)
list_of_dfs.append(df4)
list_of_dfs.append(df5)

list_of_dfs[0].head()
```

```python
df5['Gender'] == 'Male'
```

```python
df1.loc[df1.Gender=='Woman','Gender']='Female'
df2.loc[df2.Gender=='Woman','Gender']='Female'
df3.loc[df3.Gender=='Woman','Gender']='Female'
df1.loc[df1.Gender=='Man','Gender']='Male'
df2.loc[df2.Gender=='Man','Gender']='Male'
df3.loc[df3.Gender=='Man','Gender']='Male'
```

```python
only_female = df5[df5['Gender'] == 'Female']
only_male = df5[df5['Gender'] == 'Male']
only_female
```

```python
def counter(df):
    only_female = df[df['Gender'] == 'Female']
    only_male = df[df['Gender'] == 'Male']
    
    number_of_female = len(only_female)
    number_of_male = len(only_male)
    gesamt = len(df)
    
    counts = {
        'female': number_of_female,
        'male': number_of_male,
        'gesamt': gesamt
    }
    
    return counts

number = counter(df5)
    
print("Female: " + str(number['female']) + " --- Male: " + str(number['male']) + " --- Insgesamt: " + str(number['gesamt']))
```

```python
list = []

for df in list_of_dfs:
    numbers = counter(df)
    list.append(numbers)
    
print(list)
# faengt mit neuestem Datensatz an
```

```python
data = {
    'Jahr': ['2021', '2020', '2019', '2018', '2017'],
    'Frauen': [list[0]['female'], list[1]['female'], list[2]['female'], list[3]['female'], list[4]['female']],
    'Gesamtzahl': [list[0]['gesamt'], list[1]['gesamt'], list[2]['gesamt'], list[3]['gesamt'], list[4]['gesamt']] 
}
df = pandas.DataFrame(data=data)
```
