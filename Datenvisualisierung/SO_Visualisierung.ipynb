# Einlesen der Datensätze

```python
import matplotlib.pyplot as pyplot

import pandas as pandas

df1=pandas.read_csv("./stack-overflow-developer-survey-2021/survey_results_public.csv",index_col="ResponseId")
df2=pandas.read_csv("./frauenanteil_so.csv")
```

```python
df2.head()
```

# Balkendiagramm

```python
var=df2.groupby(['Jahr']).Prozent.sum()
fig=pyplot.figure()
ax=fig.add_subplot(1,1,1)
ax.set_xlabel('Jahr')
ax.set_ylabel('Anteil in Prozent')
ax.set_title('Frauenanteil im Laufe der Jahre')
my_colors=['b', 'r', 'c', 'y', 'g', 'm']
var.plot(kind='bar', color=my_colors)
pyplot.show()
```

# Liniendiagramm

```python
# Get col of a dataframe
comp=df1.ConvertedCompYearly
newdf=pandas.DataFrame(comp).sort_values(by='ConvertedCompYearly')
newdf=newdf.dropna()
newdf=newdf.reset_index(drop = True)
#Die oberen und unteren ca 5% werden abgeschnitten um eine aussagekräftigere Kurve zu erhalten.
newdf=newdf.drop(newdf.tail(2000).index)
newdf=newdf.drop(newdf.tail(2000).index)

newdf.plot(xlabel='Index', ylabel='Gehalt in Dollar', color='green')
```
